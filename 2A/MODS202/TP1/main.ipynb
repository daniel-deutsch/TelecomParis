```python
import numpy as np
import pandas as pd
import statsmodels.api as sm
import matplotlib.pyplot as plt
```

# Exercise 1

```python
df = pd.read_csv(
    "./data/wage1.raw", 
    delim_whitespace = True, 
    names = ["wage", "educ", "exper", "tenure", "nonwhite", "female", "married", "numdep", "smsa", "northcen", "south", "west", "construc", "ndurman", "trcommpu", "trade", "services", "profserv", "profocc", "clerocc", "servocc", "lwage", "expersq", "tenursq"]
)
```

# Exercise 2

```python
plt.hist(df["wage"], "auto")
```

# Exercise 3

```python
wage_mean = np.mean(df["wage"])
wage_std = np.std(df["wage"])
wage_max = np.max(df["wage"])

print(f"Mean: {wage_mean}")
print(f"Standard Deviation: {wage_std}")
print(f"Maximum: {wage_max}")
```

# Exercise 4

```python
wage_educ_cov = np.cov(df["wage"], df["educ"])
wage_educ_corr = np.corrcoef(df["wage"], df["educ"])

print("Covariance between wage and education:")
print(wage_educ_cov)
print()
print("Correlation between wage and education:")
print(wage_educ_corr)
```

# Exercise 5

```python
plt.scatter(df["educ"], df["wage"])
plt.xlabel("educ")
plt.ylabel("wage")
plt.show()
```

# Exercise 6

```python
n_females = np.sum(df["female"] == 1)
sum_females_wage = df.loc[df["female"] == 1, "wage"].sum()
mean_females_wage = sum_females_wage/n_females

print(f"The mean wage between females is {sum_females_wage}/{n_females} = {mean_females_wage}")
```

# Exercise 7

```python
n_males = np.sum(df["female"] == 0)
sum_males_wage = df.loc[df["female"] == 0, "wage"].sum()
mean_males_wage = sum_males_wage/n_males

print(f"The mean wage between males is {sum_males_wage}/{n_males} = {mean_males_wage}")
```
